## Preparing the data
In this section, we define the response variable and the different predictors. The data is scaled to 0 mean and 1 standard deviation. Here, we will assume the numerical covariates (all except ```mode``` ) are normally distributed. To scale them we will apply the following transformation:$$ Z = \frac{ X- \mu}{\sigma}$$

````python
#scale
df[df.columns[:-1]] = df[df.columns[:-1]].apply(lambda x: (x-x.mean())/x.std())

#Shuffle rows
df=df.sample(frac=1)

#response
y = df.target

#variables
X = df.drop('target', axis =1)
#train and testing data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)

````

## K- Nearest Neighbors algorithm
This a supervised clustering method which relies on the principle that *similar objects are close to one another*. An integer $k \in \mathbb N$ is picked (more on how to select $k$ later), then the distance between the object we want to classify and it's $k$ nearest neighbors is determined (Euclidean distance):

Suppose we have an $n$ dimensional feature vector:$$ \vec X = \begin{pmatrix} x_1 \\ x_2\\ .\\.\\. \\ x_n \end{pmatrix} \in \mathbb R^n$$ and one of it's $k$ - nearest neighbors $\vec X'$.

We first calculate the quantity: $$d = \sqrt{(x_1 - x_1')^2 + (x_2-x_2')^2 ... + (x_n- x_n')^2}$$ 

This distance is calculated for each of the $k$-neares

````python
model1 = KNeighborsClassifier(n_neighbors = 35)
model1.fit(X_train, y_train)
````

````python
y_pred = model1.predict(X_test)
````

````python
accuracy_score = model1.score(X_test, y_test) 
print("Accuracy score = ""{:.2f}".format(accuracy_score))
````

````python
accuracy_rate = []

for k in range(1,40):
    
    model = KNeighborsClassifier(n_neighbors=k)
    score=cross_val_score(model, X, y, cv=10)
    accuracy_rate.append(score.mean())
````

````python
error_rate = []

for k in range(1,40):
    
    model = KNeighborsClassifier(n_neighbors=k)
    score=cross_val_score(model, X, y, cv=10)
    error_rate.append(1-score.mean())
````

````python
error_rate = []

for k in range(1,40):
    
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train,y_train)
    pred_k = model.predict(X_test)
    error_rate.append(np.mean(pred_k != y_test))
````
